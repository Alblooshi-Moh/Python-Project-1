## if the population of oman is 4644384 and groth rate is 1.49 % what will be the population in 2040

```python
PPL = 4644384
GRW = 1.49 / 100
year = 2023

while year < 2040:
    PPL = PPL + PPL * GRW
    year += 1

print("The population in 2040 is:", PPL)
```

```python
PPL = 4644384
GRW = 1.49 / 100
years = 0

while years < 10:
    PPL = PPL + PPL * GRW
    years += 1

print("The population after 10 years is:", PPL)
```

```python
PPL = 4644384
GRW = 1.49 / 100
TRGPPL = 6000000
year = 2023

while PPL <= TRGPPL:
    PPL = PPL + PPL * GRW
    year += 1

print("The population exceeds 6 million in the year:", year)
```
