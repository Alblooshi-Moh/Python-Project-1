## Draw the relastionship between the expenses and the month

```python
Month = ["JAN","FEB","MARCH","APRIL"]
EXPE = [200, 300, 500, 400]
income = [500, 400, 600 ,800]
plt.plot(Month,EXPE)
plt.title("EXpenses per Month")
plt.xlabel("Month")
plt.ylabel("Expenses")
#للفصل بين الرسمينplt.show()
plt.plot(Month, income)
```
